```python
# 3.A compay will decided to give bonus to the employee according to the following criteria:
# Time period of the service -------------------> Bonus
# More then 10 Years ---------------------------> 10%
# >=6 and <=10 ---------------------------------> 8%
# <6 -------------------------------------------> 5%
# Ask the user for their salary and yers of service and then calcualte the bonus amount is, 
# and display the total salary after increment of bonus.

userSalary = int(input("Enter your current salary ₹: "))
yearExp = int(input("Enter your Experience: "))

if userSalary>15000 and yearExp>0 and yearExp<60:
    if yearExp>=10:
        bonus = "10%"
        bonusAmount = (userSalary*10)/100
        # print(bonusAmount)
        totalSalary = userSalary+bonusAmount
        print(f"Your Current Salary: {userSalary}\nTotal Experience:{yearExp}\nBonus:{bonus}\nBonusAmount ₹: {bonusAmount}\nTotal Salary ₹:{totalSalary}")
    elif yearExp>=6 and yearExp<10:
        bonus = "8%"
        bonusAmount = (userSalary*8)/100
        # print(bonusAmount)
        totalSalary = userSalary+bonusAmount
        print(f"Your Current Salary: {userSalary}\nTotal Experience:{yearExp}\nBonus:{bonus}\nBonusAmount ₹: {bonusAmount}\nTotal Salary ₹:{totalSalary}")
    elif yearExp<6:
        bonus = "5%"
        bonusAmount = (userSalary*5)/100
        # print(bonusAmount)
        totalSalary = userSalary+bonusAmount
        print(f"Your Current Salary: {userSalary}\nTotal Experience:{yearExp}\nBonus:{bonus}\nBonusAmount ₹: {bonusAmount}\nTotal Salary ₹:{totalSalary}")
else:
    print("Invalid Input")
```
