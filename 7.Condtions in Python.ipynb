```python
## 
# age range ---- Group

# 0-3  --- Baby
# 4-12 ---- child
# 13-19 --- teenager
# 20-59 ---- Adult
# 59 to upto --- senior


age = int(input("Enter your age: "))

if age>=0 and age<=3:
    print(f"You are Baby:{age} age")
elif age>=4 and age<=12:
    print(f"You are cild: {age} age")
elif age>=13 and age<=19:
    print(f"You are teenager:{age} age")
elif age>=20 and age<=59:
    print(f"You are adult:{age} age")
elif age>59:
    print(f"You are senior:{age} age")
```
