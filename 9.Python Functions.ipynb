# 2. User-defined function

Types of Function Argument 

1. Positional Arguments
2. Keyword Arguemnts
3. Default Arguments
4. Variable length Arguments

## 1.Positional Arguments

```python
def student(name,age):
    print(age,name)
    print(name,age)
student("John",78)  # Positional argument
```

# 2. Keyword Argument



```python
def greet(name,age):
    print(name,age)

greet(age=100,name="Mohan")  # (age=100,name="Mohan") keyword argument 
```

### 3.Default Arguments

```python
def greet(admin="Guest"):
    print("Current Admin:",admin)

greet("Rohit")  # positional argument
```

```python
greet()  # default argument
```

### 4. variable Length Arguments:
* ✅ *args(Non-keyword variable arguments)

```python
def total(*num):  # selector 
    print(sum(num)) # sum(value seleted)
total(10)
```

```python
total(10,12)
total(10,36,45)
total(10,-10,8,2)
```

### Scope of Variable

```python
def test():
    x1 = 10  # local variable
    print(x1)

test()
```

```python
print(x1)
```

```python
s1 = 100   # global variable
def test():
    x1 = 10  # local variable
    print("Local Variable",x1) # it works only inside the function
    print("Global Variable",s1) # its scope is inside the function and outside the function

test()
```

```python
print(x1)
```

```python
print(s1)
```

```python
# global keyword: using global keyword you can create or deine variable global of local variable
```

```python
def change():
    global x5  # create vairbale global x5
    x5 = 100  # local variable 
    x7 = "Hello world"  # local variable
    print(x7)
    print(x5)
change()
```
