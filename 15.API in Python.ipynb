## Why requests?

https://jsonplaceholder.typicode.com/users


```python
# Step-1

import requests
```

```python
help(requests)
```

```python
# Step-2

url = "https://jsonplaceholder.typicode.com/users"

# This is the API endpoint(server address)
```

```python
# step-3

response = requests.get(url)

# send request to server
# Server replies and stores result in response
```

```python
## step-4
print(response.status_code)
print(response)
```

```python
## step-5: convert response to JSON

data = response.json()
```

```python
## step: 6
print(data)
```

# Some Status Code:
* 200- Success
* 404- Not Found
* 500 - Server Error

```python
### Understanding API Response 
print(type(data))
```
