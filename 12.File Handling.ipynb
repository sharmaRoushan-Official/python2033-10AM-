### 1. What is File in Python?

A `file` is used to sotre data permanently on a disk.Python allows us to create, read, write and modify files.

Example Files:

* .txt -> Text file
* .csv -> Data file
* .json -> Structured data
* .log -> Logs
* .pdf -> Document File

Python Provides a built-in function:

```python
open("file path")
```

## 2. Syntax for opening a file

file_object = open("file Name","mode")

### Parameters


|Parameter|Description|
|---|---|
|filename|Name of the file|
|mode|Operation to perform|


### 3. File Modes

|Mode|Description|
|---|---|
|"r"|Read file(default)|
|"w"|Write file(overwrite)|
|"a"|Append Data|
|"x"|Create file|
|"rb"|Read binary|
|"wb"|Write binary|

### 5. read() Mehtod 

Read entire file content

syntax

file.read()

Example

```python
file = open("C:\\Users\\User\\Desktop\\data.txt","r")

content = file.read()
print(content)
```

## 6. read(n) Method

Read n characters only

```python
file = open("C:\\Users\\User\\Desktop\\data.txt","r")

content = file.read(12) # 12 chracters only.

print(content)
```

## 7. readline() Mthod

Reads one line at a time.

    

```python
file  = open("C:\\Users\\User\\Desktop\\data.txt","r")

line1 = file.readline() # 1st line
line2 = file.readline() # 2nd line
line3 = file.readline() # 3rd line
print(line1)
print(line2)
print(line3)
```

### 8. readlines() Method

Read all lines into a list

```python
file = open("C:\\Users\\User\\Desktop\\data.txt","r")
lines = file.readlines()
print(lines)
```

## 9. Reading Files Using Loop

Best Method for large files.

```python
file = open("C:\\Users\\User\\Desktop\\data.txt","r")

for line in file:
    print(line)

file.close()
```

```python
file.closed # True means file alread closed
```

### 10. Using with statement (Best Practice)
with atuomatically closes the file

syntax:

with open('filename','mode') as file:

```python
with open("C:\\Users\\User\\Desktop\\data.txt","r") as f:
    print(f.read())
```

```python
with open("C:\\Users\\User\\Desktop\\data.txt","r") as file:
    print(file.read())
```

```python
with open("C:\\Users\\User\\Desktop\\data.txt","r") as x:
    print(x.read())
```
