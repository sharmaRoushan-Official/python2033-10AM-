```python
# 5)Calculate the factorial of a number using a loop.

# 5! = 1x2x3x4x5 = 120
# 3! = 1x2x3 = 6
# 6! = 1x2x3x4x5x6=720
# 8! = 1x2x3x4x5x6x7x8 = 40320

n = int(input("Enter any value: "))

fact = 1 # 6 # Recursive value [Recursion]
for i in range(1,n+1):
    fact = fact * i  
print(f"The factorial of the given {n} is: {fact}")
```

```python
names = ["Ajay","Rudra","Maria","Vaishnavi","Siddharth","Bitter Guard"]
print("Original Names:",names)

i = 0
j = len(names)
# print(j)
charLen = []

totallenList = 0
while i<j:
    print(i,names[i])
    a = 0
    b = len(names[i])
    countChar = 0
    while a<b:
        print(a,names[i][a])
        countChar+=1
        a+=1
    charLen.append(countChar)
    totallenList+=1
    i+=1
print("Total Elements in the list is:",totallenList)
print("Total Characters in Your name is:",charLen)
```
