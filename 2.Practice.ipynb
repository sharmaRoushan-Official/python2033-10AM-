```python
#12. Print each characters of a string using for loop
fruits = ["apple","orange","grapes","banana","berry","PineApple"]
print("Original List",fruits)
charList = []

for fruit in fruits:
   # print(fruit)
    countChar = 0
    for ch in fruit:
        # print(ch)
        countChar+=1
    charList.append(countChar)
    # print()
print("Total charcter of elemnets in the list is:",charList)
```

```python
fruits = ["apple","orange","grapes","banana","berry","PineApple"]
print("Original List",fruits)
charList = []

i = 0
while i<len(fruits):
    print(i,fruits[i])
    j = 0
    countChar = 0
    while j<len(fruits[i]):
        print(j,fruits[i][j])
        countChar+=1
        j+=1
    charList.append(countChar)
    # print(f"Total Charcters of fruits {fruits[i]}:{countChar}")
    i+=1
print("Total character of elements in the list is:",charList)
```

```python

vowels = "aeiou"
str1 = "hello WORld"

i = 0
vowChar = []
vowCount = 0
consChar = []
consCount = 0
while i<len(str1)-1:
    print(i,str1[i])
    i+=1
    if str1[i].lower() in vowels:
        vowChar.append(str1[i])
        vowCount+=1
    else:
        if str1[i] == " ":
            continue
        consChar.append(str1[i])
        consCount+=1
    
print("Vowel characters list:",vowChar,"=",vowCount)
print("Consonent characters list:",consChar,"=",consCount)
```

```python
#15.Reverse a string using a for loop.
# for char in reversed("python"):
#     print(char)

x = "Python"
print("Original String:",x)
revString = ""
for i in x:
    revString = i+revString
    # print(i)
print("Reversed String:",revString)
```

```python
# while loop

x = "Python"
print("Original String:",x)

i = 0
revString = ""
while i<len(x):
    # print(i)
    revString = x[i]+revString
    i+=1
print("Reversed String:",revString)
```
