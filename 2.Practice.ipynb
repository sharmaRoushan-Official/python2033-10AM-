```python
## 6. Print multiplication tables from 1 to 10

for i in range(1,11):
    # print(i)
    print(f"========Table of {i}==========")
    for j in range(1,11):
        print(f"{i}x{j}={i*j}")
    print()
```

```python
## 

i=1
while i<11:
    print(f"===========Table of {i}=============")
    j = 1
    while j<11:
        print(f"{i}x{j}={i*j}")
        j+=1
    print()
    i+=1
```

```python
# 7. find the sum of the numbers from 1 to 100

sumVal = 0
for i in range(1,101):
    sumVal = sumVal+i
    # print(i)
print(f"Sum of the value from 1-100 is:",sumVal)
```

```python
## while loop

i = 1
sumValue=0
while i<101:
    # print(i)
    sumValue = sumValue+i
    i+=1
print(f"Sum of the value from 1-100 is:",sumValue)
```

```python
## 8.find the sum of even numbers from 1 to 100

listEven = []
sumEven=0
for i in range(0,101):
    if i%2==0:
        # print(i)
        listEven.append(i)
        sumEven+=i
print("List Even:",listEven,"=",sumEven)
```

```python
### while loop

i=0

listEven = []
sumEven = 0
while i<101:
    # print(i)
    if i%2==0:
        # print(i)
        listEven.append(i)
        sumEven = sumEven+i
    i+=1
print("Even List:",listEven)
print("Sum of Even:",sumEven)
```
