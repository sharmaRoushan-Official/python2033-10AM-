### 6. Dictionary in Python

* A `Dictionary` in Python is buit-in data type used to store data in the form of `key-value` pairs.

* It is one of the most commonly used in data structures because it allows fast searching, updating, and retrieving values.


### Definition of Dictionary

A dictionary is written using curly braces {} and contains elements in the format:
key:value

```python
# display empty dictionary

x = {}
print(x,type(x))
```

```python
## empty set
x1 = set()
print(x1,type(x1))
```

```python
x1 = {1}
print(x1,type(x1))
```

```python
## Dictionary 

students = {
    "name":"ajay",
    "age":21,
    "course":"Python"
}
print(students)
print(type(students))
```

```python
students = {"name":"ajay","age":21,"course":"Python"}
print(students)
print(type(students))
```

## 2. Key Features of Dictionary 

✅ 1. Key-Value pair Structure

Each item is stored as:
key:value

✅ 2. Key Must Be Unique
If you repeat a key, the lates value is overwrite the old value.

```python
data = {"a":10,"a":20,"a":30}
print(data)
```

✅ 3. Value Can be Duplicate

value can repeat

```python
marks = {"Math":90,"Science":90}
print(marks)
```

✅ 4. Dictionary is Mutable

You can modify dictionary elements after creation.

✅ 5. Dictionaries are Unordered(but insertion ordered in Python 3.7+)

Python 3.7+ maintains insertion order, but dictionary is still treated as unordered mapping.

```python
import sys
print(sys.version)
```

✅ 3. Creating a Dictinary 

Method 1: Using curly braces {}

```python
emp = {"id":101,"name":"Ravi","salary":45000}
print(emp)
print(type(emp))
```

Method 2: Uisng dict() constructor

```python
emp = dict(id=101,name="Ravi",salary=45000)
print(emp)
print(type(emp))
```

Method 3: Creating an empty dictionary

```python
d = {}
print(d,type(d))
```

```python
d = dict()
print(d,type(d))
```

### Dictinary with Mixed Data types



```python
data = {
    "name":"Ravi",  # string
    "age":22,  # integer
    "marks":[90,85,45], # list
    "passed":True # boolean
}
print(data)
```

```python
print(type(data['name']))
print(type(data['age']))
print(type(data['marks']))
print(type(data['passed']))
```

### 11. Nested Dictionary 

A dictionary inside another dictionary is called nested dictonary.

```python
students = {
    101:{
        'name':"Arun",
        "age":21
    },
    102:{
    'name':'Bablu',
    "age":22
    }
}
students
```

```python
students = {
    101:{'name':"Arun","age":21},
    102:{'name':'Bablu',"age":22}
}
students
```

### Dictionary VS List
---
|Feature|List|Dictionary|
|---|---|---|
|Stored As|Values only|Key-Value|
|Indexing|By index number|by key|
|Dupicate allowed|Yes|key no, values yes|
|Ordered|Yes|Insertion ordered(3.7+)|
|Search Speed| slower|faster|
