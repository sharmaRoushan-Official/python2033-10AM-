## 1.  for Loop 

⭐ 1. What is for Loop

A for loop is used to iterate(travese) over a sequence and execute a set of statments repeatedly.

```python
for i in [1,2,3]:
    print(i)
```

Explanation:

* for -> keyword
* variable -> takes each varlue one by one
* in -> keyword to access elements from sequence
* sequence -> list, tuple, string, range, dict,set
* : - end of the line
* Indentation -> defines loop block

```python
### 3. Flow of Execution

for x in [10,20,30]:
    print(x)
```

#### Flow:

1. x = 10 -> print(10)
2. x = 20 -> print(20)
3. x = 30 -> print(30)

### 12. for loop with Dictionary



```python
d = {'a':'apple',"b":'boy',"c":'cat','d':10}

for key in d:
    print(key)
```

```python
for value in d.values():
    print(value)
```

```python
for key,value in d.items():
    print(key,value)
```

```python
for key,value in d.items():
    print(key,":",value)
```

### 13. Using enumerate() in for loop

Used when you need index+value

```python
names = ['ajay',"vaishnavi","Rudra","maria"]

for i in names:
    print(i)
```

```python
for index,name in enumerate(names):
    print(index,name)
```

Used for matrices,patterns,tables
