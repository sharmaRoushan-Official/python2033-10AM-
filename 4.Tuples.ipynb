### 4. Tuples 

### ⭐ Tuples is Immutable

That means once a tuple is created, its elements cannot be changed at any cost.

```python
t = (10,20,30)
print(t)
print(type(t))
```

```python
x = 10,
print(x,type(x))
```

```python
x1 = "hello world",
print(x1,type(x1))
```

```python
x = ("hello world",10,True)
print(x,type(x))
```

## Tuple packing and Unpacking



```python
# packing
t = 1,2,3
print(t)
```

```python
# unpacking

t = (1, 2, 3)
a,b,c = t
print(a)
print(b)
print(c)
```

```python
list1 = ["Siddharth","Ajay","Vaishnavi",10.25,True,False,"Maria","Rudra"]
print(list1)
```

```python
x1 = tuple(list1)
print(x1)
```

## 5. Tuple Indexing: Positive and Negative Indexing


```python
# Positive Indexing
t = (5,10,15,20)
print(t[0])
print(t[1])
print(t[2])
print(t[3])
```

```python
# Negative Indexing

t = (5,10,15,20)
print(t[-1])
print(t[-2])
print(t[-3])
print(t[-4])
```

### Tuple slicing



```python
t = (5,10,15,20,40,50,60,70,-80)

# sytax:
# [starts:ends] 
# [starts:ends:step]

t[3:6]
```

```python
## Striding Value 
print(t)
t[::2]
```

```python
t[::3]
```

### 8. Tuple is Immtable (cannot Modify)

```python
# updating the value of tuple which is 0 position
t = (10,20,30)
t[0] = 100
t
```

### So you cannot:
 * add new element
 * update element
 * delete element

```python
# adding the value of tuple which is 0 position
t = (10,20,30)
t.append(20)
print(t)
```

```python
# updating the value of tuple which is 0 position
t = (10,20,30)
t.pop()
```

```python
# delete the value of tuple which is 0 position
t = (10,20,30)
del(t[0])
```

```python
### Tuple with Different Data Types

t = (10,"Python",3.5,True, False,(10,20),[20,50],{1,1,1,2,2,2,33})
print(t)
```
