### 5. Sets in Python


In Python, a `set` is a built-in data type used to store `multiple unique values` in a single variable.


A set is mainly used when you want:

* ✅ No duplicate values
* ✅ Fast searching (membership checking)
* ✅ Mathematical set operations like `Union, Intersection, Difference and Symmetric Difference`


Define:
 * A __set__ is an unordered collection of elements.

   * KeyPoints:
   * __Unordered__ -> elements have no fixed position.
   * __Unique__ -> Duplicates are automatically removed.
   * __Mutable__ -> You can add or remove elements
   * Written using {} braces.

```python
# () -> parentheses - tuples
# [] -> List
# {} -> set
# {} -> dictionary
```

```python
# Operations in Set

# Union, Intersection, Difference, Symmetric Difference
```

```python
## Union-> It returns unique value of two sets
s1 = {"Ravi","Kiran"}
s2 = {"Suresh","Kiran"}

print(s1.union(s2))
```

```python
## Intersection :- It remove unique value and return duplicate value
s1 = {"Ravi","Kiran"}
s2 = {"Suresh","Kiran"}

print(s1.intersection(s2))
```

```python
s1&s2
```

```python
# Difference
s1 = {"Ravi","Kiran"} # a
s2 = {"Suresh","Kiran"} # b

print(s1.difference(s2))
print(s2.difference(s1))
```

### ✅ When To Use Set?

* You Want to store unique values.
* You want to remove duplicate quickly.
* You want to check membership value fast(in or not in).
* You want to perform operation on set items(union, intersection, difference, symmetric_difference)
