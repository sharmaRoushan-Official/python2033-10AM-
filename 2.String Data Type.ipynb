````python
data = "hello WORld"
print("Original String:",data)

# upper()
# Important String Method(Very Important Method)

# upper()
# lower()
# capitalize()
# title()
# replace()
# swapcase()

data2 = data.upper()
print(data2)


# lower()
data3 = data.lower()
print(data3)


# capitalize()
data4 = data.capitalize()
print(data4)

#title()
data5 = data.title()
print(data5)


data6 = data.replace("WORld","Ajay")
print(data6)

data7 = data.swapcase()
print(data7)
````

| Character | Description            | Example         | Output     |   |

| `.`       | Dot (lists)            | `1\. item`      | 1. item    |   |
| `!`       | Image syntax           | `\!important`   | !important |   |
