# 1. What is String?

* A `String` in Python is a Sequence of Unicode characters enclosed in Quotes.

```python
s1 = "Hello world"

s2 = 'Hello Ajay'

s3 = """Multi-line string"""   # ''' same as it is ''' # docstring 

# char x[20] = "Hello world";

print(s1)
print(s2)
print(s3)
```

### Key Points:

 * __String are immutable (cannot be changed after creation)__
 * __Index(Ordered Collections)__
 * __Positive Indexed starts from [0] and Negative Indexed starts [-1]__
 * __Supports slicing, iteration, and many buit-in methods.__

```python
s= 'Python'
print(s,type(s))
```

```python
# Positive Indexing
print(s[0])
print(s[1])
print(s[2])
print(s[3])
print(s[4])
print(s[5])
```

```python
# Negative Indexing
print(s[-1])
print(s[-2])
print(s[-3])
print(s[-4])
print(s[-5])
print(s[-6])
```
