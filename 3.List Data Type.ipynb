### 8. List Methods 


|__Method__|__Work__|
|---|---|
|`append(x)`| Add item at end|
|`extend(iterable)`|Add multiple value|
|`insert(i,x)`| Insert at index|
|`remove(x)`|Remove first occurrence|
|`pop()`|Pop last element|
|`pop(i)`| Remove item at index|
|`clear()`| Remove all items|
|`index(x)`| Return index value|
|`count(x)`| Counts occurrence|
|`sort()`| Sort list [ascending]|
|`reverse()`|Reverse the list|
|`copy()`|Returns shallow copy|
