### 3. List Data Type 

A `list` in Python is a mutable(changeable), ordered, indexed collections that can store multiple values(including mixed data types). Lists are one of the most used data types in Python.

```python
myList = []   # empyt list
print(myList)
print(type(myList))
```

```python
str1 = ''
print(str1)
print(type(str1))
```

```python
myList = [1,2,3,4,5]
print(myList)
```

```python
myList1 = [10,20,30,40,50]
print(myList1)
```

```python
# s[0]  ---> Accessing address value of the variable [it is not list]
```

```python
# Positive Indexing
myList1 = [10,20,30,40,50]
print(myList1)


print(myList1[0])
print(myList1[1])
print(myList1[2])
print(myList1[3])
print(myList1[4])
```

```python
# Negative Indexing
print(myList1[-1])
print(myList1[-2])
print(myList1[-3])
print(myList1[-4])
print(myList1[-5])
```

### ✅ 2. Features of List

|__Feature__|__Explanation__|
|---|---|
|`ordered`|Items maintain insertion ordered|
|`Indexed`|Each element has an indexed|
|`Mutable`|Can Modify after create list|
|`Allow duplicates`|Same value can repeat|
|`Mixed type allowed`|int,float, str,bool, object, tuple, list,, dictionary, set etc.|

## 4. List Slicing

```python
x = [10,20,30,40,50,60,70,80,90]

x[3:6]
```

```python
x[-1]
```

```python
x[8:]
```

4. List Slicing
syntax: [start:end:step]

```python
x = [10,20,30,40,50,60,70,80,90]    # [10,40,70]
 
x
```

```python
x[0:9:3]
```

```python
x[::3]
```
